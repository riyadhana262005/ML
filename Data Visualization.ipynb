# Relationship plot

```python
ss.relplot(x='total_bill',y='tip', data=tip,  col="time", hue="day",style="day", kind="scatter")
```

```python
ss.relplot(x='total_bill',y='tip', data=tip,  col="time", hue="day",style="day", kind="line")
```

# Categorigal Box plot

```python
ss.catplot(x='size', y='total_bill', kind='box', data=tip)
```

# Regression plot
