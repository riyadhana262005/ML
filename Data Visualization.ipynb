# Creating data

```python
movies = ['joe','vip','remo','vvs']
ratings = [10,9,8,10]
```

# importing libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# scatter plot

```python
plt.scatter(movies,ratings)
```

# Line plot

```python
plt.plot(movies,ratings, linestyle='dashdot')
```

# pie chart

```python
plt.pie(ratings, labels=movies)
```

# Bar chat

```python
plt.bar(movies,ratings)
```

```python
plt.bar(ratings,movies)
```
